```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
for i in df.columns:
  if df[i].dtype == 'object':
    print(df[i].value_counts())
    print('************************************')
```

```python
df=df.replace({"t": True, "f": False})
```

```python
df=df.replace({'?': np.nan})
```

```python
df=df.replace({"M": 1, "F": 0})
```

```python
del df['TBG measured']
```

```python
# too many missing values
del df['TBG']
```

```python
# remove all rows with missing values
df = df.dropna()
```

```python
# hyperthyroid conditions:
# 	A	hyperthyroid
# 	B	T3 toxic
# 	C	toxic goitre
# 	D	secondary toxic

# hypothyroid conditions:
# 	E	hypothyroid
# 	F	primary hypothyroid
# 	G	compensated hypothyroid
# 	H	secondary hypothyroid

hyper_cond = ['A', 'B', 'C', 'D']
hypo_cond = ['E', 'F', 'G', 'H']
healthy_cond = ['-']

for idx, row in df.iterrows():
    cond = row['diagnosis']
    if cond in hyper_cond:
        row = row.replace({cond: 'hyperthyroid'})
    elif cond in hypo_cond:
        row = row.replace({cond: 'hypothyroid'})
    elif cond in healthy_cond:
        row = row.replace({cond: 'healthy'})
    else:
        df = df.drop(idx)
```

```python
for i in df.columns:
  if df[i].dtype == 'object':
    print(df[i].value_counts())
    print('************************************')
```
